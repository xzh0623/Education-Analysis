```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("categorical_dataset.csv")

# 绘制散点图
plt.scatter(data['SALARY'], data['AGE'])
plt.xlabel('SALARY')
plt.ylabel('AGE')
plt.title('Scatter Plot of SALARY and AGE')
plt.show()

# 查看组合中有多少人
num_people = data.shape[0]
print(f"组合中有 {num_people} 人。")
```
