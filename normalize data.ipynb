```python
import pandas as pd
numeric_data = pd.read_csv("./dataset/numerical_dataset.csv")
```

```python
numeric_data.columns[1:]
```

```python
from sklearn.preprocessing import StandardScaler

columns_to_normalize = numeric_data.columns[1:]
scaler = StandardScaler()
numeric_data[columns_to_normalize] = scaler.fit_transform(numeric_data[columns_to_normalize])
print(numeric_data)
```

```python
categorical_data = pd.read_csv("./dataset/categorical_dataset.csv")
```

```python
categorical_data.columns[1:]
```

```python
import pandas as pd

columns_to_encode = categorical_data.columns[1:]
categorical_data_encoded = pd.get_dummies(categorical_data, columns=columns_to_encode)
print(categorical_data_encoded)
```

```python
categorical_data_encoded
```

```python
merged_data = pd.merge(categorical_data_encoded, numeric_data, on='STUDENTID')
```

```python
merged_data.to_csv("./dataset/normalized_dataset.csv",index=False)
```
